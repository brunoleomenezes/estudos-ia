```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model  
from mlxtend.plotting import plot_decision_regions
import pandas as pd
```

```python
modelo =  linear_model.LinearRegression()
carros = pd.read_csv('/content/mt_cars.csv', sep=",").astype(int)
carros
```

```python
#converter para numpy
carros = carros.to_numpy()
```

```python
#prevendo o número de carburadores
y = carros[:,9]
#consumo e cilindoros
X = carros[:, 0:2]
X
```

```python
y
```

```python
modelo.fit(X, y)
```

```python
fig = plot_decision_regions(X=X, y=y,clf=modelo)
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left", ncol=2)
plt.title('Limites de Decisão')
plt.xlabel('Consumo')
plt.ylabel('Cilindros')
plt.show()
```
