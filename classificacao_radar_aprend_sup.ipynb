# Quais características (features) das aeronaves? 
# R = Civil, motorização e hélice.

# x -> civil + mortorização + hélice.
# y -> 0 rotativas ou 1 fixas.

# Uma aeronave que não é civil, tem morotização e não tem hélice. Seria de asas fixas ou rotativas? 
# Existe uma margem de erro, ou seja, existe taxa de acerto (ex: acurácia).

```python
# Features (1 sim, 0 não)
# Civil?
# Motorização?
# Hélice?
```

```python
# Três itens com três características (features)
fixa1 = [0, 1, 0]
fixa2 = [0, 1, 1]
fixa3 = [1, 1, 0]

rotativa1 = [0, 1, 1]
rotativa2 = [1, 0, 1]
rotativa3 = [1, 1, 1]
```

```python
# 1 -> fixa, 0 -> rottiva
dados = [fixa1, fixa2, fixa3, rotativa1, rotativa2, rotativa3]
classes = [1,1,1,0,0,0]
# f(x) = y onde x = dados e y = classes ou seja dados (x) de treino e classes (y) de treino. 
# As classes também são chamadas de etiquetas (labels).
```

```python
from sklearn.svm import LinearSVC
model = LinearSVC()
# vamos ensinar ao nosso sistema o que é rotativa ou fixa
# vamos colocar (fit) os dados dentro do modelo 
model.fit(dados, classes)
```

```python
aeronave_misteriosa = [1,1,1]
```

```python
# modelo realize a previsão da aeronave_misteriosa
model.predict([aeronave_misteriosa])
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
```

```python
teste = [misterio1, misterio2, misterio3] # teste = x
```

```python
previsoes = model.predict(teste)
previsoes
# 1 -> fixa, 0 -> rotativa
```

#Errado! O primeiro era rotativa e os demais eram fixas ([0, 1, 1]). 

```python
1/3
```

```python
testes_classes = [0, 1, 1] # teste = y
```

```python
previsoes == testes_classes
```

```python
corretos = (previsoes == testes_classes).sum()
corretos
```

```python
total = len(teste)
taxa_de_acerto = corretos/total
print("Taxa de acerto %.4f" % (taxa_de_acerto * 100))
```

```python
# Taxa de acerto - Acurácia
from sklearn.metrics import accuracy_score
taxa_de_acerto - accuracy_score(testes_classes, previsoes)
print("Taxa de acerto %.4f" % (taxa_de_acerto * 100))
```
