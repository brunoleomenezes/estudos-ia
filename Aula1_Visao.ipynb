```python
img_cinza = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
classif = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
rostos = classif.detectMultiScale(img_cinza, 1.2 , 5) #piradimização
len(rostos)
```

```python
rostos[0] # x e y com altura e comprimento
```

```python
imganot = img.copy()
for (x,y,comp,alt) in rostos:
    cv2.rectangle(imganot, (x,y), (x+comp, y+alt), (250, 250, 0), 3)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(imganot)
```

```python
rosto_img = 0

for (x,y,comp,alt) in rostos:
    rosto_img += 1
    img_recort2 = cv2.cvtColor(img[y:y+comp, x:x+alt], cv2.COLOR_RGB2BGR)
    cv2.imwrite("rosto_" + str(rosto_img) + ".png", img_recort2)
```
