```python
# Importando bibliotecas adicionais para análise e visualização
import numpy as np
from math import radians, cos, sin, asin, sqrt
import matplotlib.pyplot as plt

# Função para calcular a distância entre dois pontos usando a fórmula Haversine
def haversine(lon1, lat1, lon2, lat2):
    # Convertendo coordenadas de graus para radianos
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # Fórmula Haversine
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Raio da Terra em quilômetros
    return c * r

# Calculando as distâncias entre cada par de localizações
for i in range(len(df)):
    for j in range(i+1, len(df)):
        dist = haversine(df.iloc[i]['longitude'], df.iloc[i]['latitude'],
                         df.iloc[j]['longitude'], df.iloc[j]['latitude'])
        print(f"Distância entre {df.iloc[i]['location']} e {df.iloc[j]['location']}: {dist:.2f} km")

# Determinando o ponto central (média das latitudes e longitudes)
central_latitude = df['latitude'].mean()
central_longitude = df['longitude'].mean()
print(f"\nPonto central: Latitude {central_latitude}, Longitude {central_longitude}")

# Visualização simples das localizações
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], color='blue')
plt.scatter(central_longitude, central_latitude, color='red', label='Ponto Central')
plt.title('Localizações e Ponto Central')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

```python
import pandas as pd
from itertools import permutations
from haversine import haversine

# Exemplo de dados: 5 pontos em Resende, RJ
data = {
    'Ponto': ['A', 'B', 'C', 'D', 'E'],
    'Latitude': [-22.4695, -22.4704, -22.4713, -22.4722, -22.4731],
    'Longitude': [-44.4467, -44.4476, -44.4485, -44.4494, -44.4503]
}

df = pd.DataFrame(data)

# Função para calcular a distância total de uma rota
def calcular_distancia_total(rota):
    distancia_total = 0
    for i in range(len(rota) - 1):
        ponto_atual = (df.loc[df['Ponto'] == rota[i], 'Latitude'].values[0],
                       df.loc[df['Ponto'] == rota[i], 'Longitude'].values[0])
        proximo_ponto = (df.loc[df['Ponto'] == rota[i + 1], 'Latitude'].values[0],
                         df.loc[df['Ponto'] == rota[i + 1], 'Longitude'].values[0])
        distancia_total += haversine(ponto_atual, proximo_ponto)
    return distancia_total

# Gerando todas as permutações possíveis de rotas (ignorando a complexidade computacional para este exemplo)
rotas = permutations(df['Ponto'])

# Encontrando a rota com a menor distância total
rota_otimizada = min(rotas, key=calcular_distancia_total)
distancia_otimizada = calcular_distancia_total(rota_otimizada)

print(f"A rota otimizada é: {rota_otimizada} com uma distância total de {distancia_otimizada:.2f} km")
```

```python
import gmplot

# Latitude e Longitude iniciais para o centro do mapa
latitude_centro = df['Latitude'].mean()
longitude_centro = df['Longitude'].mean()

# Crie um objeto gmplot
gmap = gmplot.GoogleMapPlotter(latitude_centro, longitude_centro, 13)

# Coordenadas para a rota otimizada
lats, lngs = [], []
for ponto in rota_otimizada:
    lats.append(df.loc[df['Ponto'] == ponto, 'Latitude'].values[0])
    lngs.append(df.loc[df['Ponto'] == ponto, 'Longitude'].values[0])

# Adicione a rota ao mapa
gmap.plot(lats, lngs, 'cornflowerblue', edge_width=3.0)

# Adicione marcadores para cada ponto
for lat, lng, ponto in zip(lats, lngs, rota_otimizada):
    gmap.marker(lat, lng, title=ponto)

# Salve o mapa em um arquivo HTML
gmap.draw("rota_otimizada_resende.html")
```
