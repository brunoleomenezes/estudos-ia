```python
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Acesando imagem
arquivo_imagem = plt.imread('/content/drive/MyDrive/Curso_Visao/post_thumbnail-d8c7a3dfbb8aef187db248252733f572.jpeg')
```

```python
sensor_de_faces = MTCNN()
# detect faces in the image
resultados = sensor_de_faces.detect_faces(arquivo_imagem)
```

```python
resultados
```

```python
# draw an image with detected objects
def retangulo(nome_do_arquivo, lista_com_resultados):
    # acesso a imagem
    dado = plt.imread(nome_do_arquivo)
    # plotando 
    plt.imshow(dado)
    # retangulos
    by = plt.gca()
    # plotando 2
    for resultado in lista_com_resultados:
        x, y, largura, altura = resultado['box']
        retan = plt.Rectangle((x, y), largura, altura, fill=False, color='blue')
        by.add_patch(retan)
    plt.show()
```

```python
retangulo('/content/drive/MyDrive/Curso_Visao/post_thumbnail-d8c7a3dfbb8aef187db248252733f572.jpeg', resultados)
```

```python
# draw an image with detected objects
def retangulo_and_keypoints(nome_do_arquivo, lista_com_resultados):
    dado = plt.imread(nome_do_arquivo)
    # plot the image
    plt.imshow(dado)
    # get the context for drawing boxes
    by = plt.gca()
    # plot each box
    for resultado in lista_com_resultados:
        x, y, largura, altura = resultado['box']
        retan = plt.Rectangle((x, y), largura, altura,fill=False, color='blue')
        by.add_patch(retan)
        # pontos
        for key, valor in resultado['keypoints'].items():
            ponto = plt.Circle(valor, radius=1, color='yellow')
            by.add_patch(ponto)
    plt.show()
```

```python
retangulo_and_keypoints('/content/drive/MyDrive/Curso_Visao/post_thumbnail-d8c7a3dfbb8aef187db248252733f572.jpeg', resultados)
```
