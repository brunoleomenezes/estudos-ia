```python
from sklearn.feature_extraction.text import CountVectorizer

duas_frases = ["Assisti um filme ótimo", "Assisti um filme ruim"]
vet = CountVectorizer(lowercase=False)
bw= vet.fit_transform(duas_frases)
vet.get_feature_names_out()
```

```python
matriz = pd.DataFrame.sparse.from_spmatrix(bw, columns=vet.get_feature_names_out())
matriz
```

```python
vet = CountVectorizer(lowercase=False)
bw = vet.fit_transform(textos.text_pt)
print(bw.shape) #Dimensões
```

```python
vet = CountVectorizer(lowercase=False, max_features=50) #Palavras que mais aparecem no tamanho desejado
bw = vet.fit_transform(textos.text_pt)
print(bw.shape) #Dimensões
```

```python
from sklearn.model_selection import train_test_split

treino, teste, c_treino, c_teste = train_test_split(bw, textos.sentiment, random_state = 42)

r_log = LogisticRegression()
r_log.fit(treino, c_treino)
ac = r_log.score(teste, c_teste)
print(ac) #Previsão quase 66%
```

```python
def c_t(texto, coluna_texto, coluna_classificacao):
    vet = CountVectorizer(lowercase=False, max_features=50)
    bw = vet.fit_transform(textos.text_pt)
    treino, teste, c_treino, c_teste = train_test_split(bw, textos.sentiment, random_state = 42)

    r_log = LogisticRegression()
    r_log.fit(treino, c_treino)
    return r_log.score(teste, c_teste)
print(c_t(textos, "text_pt", "classificacao"))
```

#Como melhorar?

```python
%matplotlib inline

from wordcloud import WordCloud
import matplotlib.pyplot as plt

tudo = ' '.join([texto for texto in textos.text_pt])

nuvem = WordCloud(width= 800, height= 500, max_font_size = 110, collocations = False).generate(tudo)
len(tudo)

plt.figure(figsize=(10,7))
plt.imshow(nuvem, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
col_text = textos.text_pt
```

```python
def nuvem_neg(textos, col_text):
    negativo = textos.query("sentiment == 'neg'")
    tudo = ' '.join([texto for texto in negativo[col_text]])
    nuvem = WordCloud(width= 800, height= 500, max_font_size = 110, collocations = False).generate(tudo)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
nuvem_neg(textos, "text_pt")
```

```python
def nuvem_pos(textos, col_text):
    positivo = textos.query("sentiment == 'pos'")
    tudo = ' '.join([texto for texto in positivo[col_text]])
    nuvem = WordCloud(width= 800, height= 500, max_font_size = 110, collocations = False).generate(tudo)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
nuvem_pos(textos, "text_pt")
```
