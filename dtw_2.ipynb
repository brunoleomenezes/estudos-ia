```python
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
import random
import numpy as np
```

```python
x = np.arange(0, 20, .5)
serie_1 = np.sin(x)
serie_2 = np.sin(x - 1)
path = dtw.warping_path(serie_1, serie_2)
dtwvis.plot_warping(serie_1, serie_2, path)
distance = dtw.distance(serie_1, serie_2)
```

```python
d, paths = dtw.warping_paths(serie_1, serie_2, window=21)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(serie_1, serie_2, paths, best_path)
```
