```python
import numpy as np
import matplotlib.pyplot as plt

# equação da reta ax + by + c = 0
a = -1
b = 4
c = 0.4

x = np.linspace(-2, 4, 50) #intervalo e quantidades de valores do intervalo
print(x)

# isolando y = (-a*x -c)/b
y = (-a*x -c)/b

# plotando eixo x e y 
plt.axvline(0, -1, 1, color='k', linewidth=1)
plt.axhline(0, -2, 4, color='k', linewidth=1)

# plotando reta
plt.plot(x, y)
plt.grid(True)
```

#O mesmo código acima, apenas implementa uma função para resolver diferentes pontos na reta.

```python
import numpy as np
import matplotlib.pyplot as plt

a = -1
b = 4
c = 0.4

def plotline(a, b, c): 
  x = np.linspace(-2, 4, 50) 
  y = (-a*x -c)/b

  plt.axvline(0, -1, 1, color='k', linewidth=1)
  plt.axhline(0, -2, 4, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)
  
plotline(a, b, c)
```

```python
p1 = (2, 0.4) # ponto 1 com x = 2 e y = 0.4
p2 = (1, 0.6)
p3 = (3, -0.4)

plotline(a,b,c)
plt.plot(p1[0], p1[1], color='b', marker='o')
plt.plot(p2[0], p2[1], color='r', marker='o')
plt.plot(p3[0], p3[1], color='g', marker='o')
```

```python
ret1 = a*p1[0] + b*p1[1] + c  #ax + by + c = 0
print("%.2f" % ret1) #nulo -> f(x) = 0

ret2 = a*p2[0] + b*p2[1] + c
print("%.2f" % ret2) #positivo -> f(x) > 0

ret3 = a*p3[0] + b*p3[1] + c
print("%.2f" % ret3) #negativo -> f(x) < 0
```
